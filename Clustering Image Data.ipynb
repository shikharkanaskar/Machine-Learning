```python
import numpy as np
from skimage import io
import pandas as pd
from sklearn.cluster import KMeans
```

### 1. Reading the image

```python
dog = io.imread('dogs.jpeg')
io.imshow(dog)
```

### 2. Checking dimension of the currrent image and  converting to a 2D array

```python
print("Current shape of the array -->", dog.shape)
print("Desired shape --> (",(dog.shape[0]*dog.shape[1]),",", dog.shape[2],")")
```

```python
dog_reshaped = dog.reshape((dog.shape[0]*dog.shape[1]), dog.shape[2])
```

```python
print("New shape of the array -->", dog_reshaped.shape)
```

The array is converted to the desired shape and now we can use kmeans clustering on this new array

### 3. Using K Means clustering

```python
kmeans = KMeans(n_clusters = 3,random_state=0,n_init=200).fit(dog_reshaped)
```

```python
io.imshow(kmeans.labels_.reshape(dog.shape[0],dog.shape[1]))
```
